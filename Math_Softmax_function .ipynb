```python
# import libraries
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
     
```

```python
# the list of numbers
z = [1,2,3]
```

```python
# compute the softmax result
num = np.exp(z)
den = np.sum( np.exp(z) )
sigma = num / den
```

```python
sigma
```

```python
np.sum(sigma)
```

```python
# repeating with some random integers 
z =np.random.randint(-5, high=15, size= 25)
z
```

```python
# Computing the softmax result 
num = np.exp(z)
den = np.sum(num)
sigma = num/den
```

```python
sigma
```

```python
# Comparing 
plt.plot(z,sigma,'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmaxified ro ')
plt.yscale('log')
plt.title(' summation ro = %g' %np.sum(sigma))
plt.show()
```

## slightly more involved using torch.nn


```python
softfun = nn.Softmax(dim = 0)

# then apply the data to that function 
sigmaT = softfun(torch.Tensor(z))

# results 
print(sigmaT)
```

```python
# show that they are the same
plt.plot(sigma,sigmaT,'ko')
plt.xlabel('"Manual" softmax')
plt.ylabel('Pytorch nn.Softmax')
plt.title(f'The two methods correlate at r={np.corrcoef(sigma,sigmaT)[0,1]}')
plt.show()
```
